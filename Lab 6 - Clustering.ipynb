    4. Relevant Information Paragraph:

       Car Evaluation Database was derived from a simple hierarchical
       decision model originally developed for the demonstration of DEX
       (M. Bohanec, V. Rajkovic: Expert system for decision
       making. Sistemica 1(1), pp. 145-157, 1990.). The model evaluates
       cars according to the following concept structure:

       CAR                      car acceptability
       . PRICE                  overall price
       . . buying               buying price
       . . maint                price of the maintenance
       . TECH                   technical characteristics
       . . COMFORT              comfort
       . . . doors              number of doors
       . . . persons            capacity in terms of persons to carry
       . . . lug_boot           the size of luggage boot
       . . safety               estimated safety of the car

       Input attributes are printed in lowercase. Besides the target
       concept (CAR), the model includes three intermediate concepts:
       PRICE, TECH, COMFORT. Every concept is in the original model
       related to its lower level descendants by a set of examples (for
       these examples sets see http://www-ai.ijs.si/BlazZupan/car.html).

       The Car Evaluation Database contains examples with the structural
       information removed, i.e., directly relates CAR to the six input
       attributes: buying, maint, doors, persons, lug_boot, safety.

       Because of known underlying concept structure, this database may be
       particularly useful for testing constructive induction and
       structure discovery methods.

```python
def ordinal_encoding(values):
    unique_values = list(set(values))
    encoding = {}
    for i, value in enumerate(sorted(unique_values)):
        encoding[value] = i
    encoded_values = [encoding[value] for value in values]
    return encoded_values
```

```python
for col in df.columns:
    df[col] = ordinal_encoding(df[col])
```

### Question 2

```python
sse = []
K = range(1, 7)
for k in K:
    kmeans = KMeans(n_clusters=k, init='k-means++')
    kmeans.fit(df)
    sse.append(kmeans.inertia_)

plt.plot(K, sse, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.title('The Elbow Method')
plt.show()
```

```python
sil = []
K = range(2, 7)
for k in K:
    kmeans = KMeans(n_clusters=k, init='k-means++')
    kmeans.fit(df)
    sil.append(silhouette_score(df, kmeans.labels_))

plt.plot(K, sil, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('The Silhouette Method')
plt.show()
```

**Observation:**<br>
- As Per Elbow Method Optimal Number of Clusters can be 2,3,4
- After applying Silhouette Method we are able to find the correct Number of Clusters i.e. 2,4.

### Question 3

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

for i in range(2,7):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans_labels = kmeans.fit_predict(df_scaled)
    kmeans_silhouette = silhouette_score(df_scaled, kmeans_labels)
    print("KMeans silhouette score:", kmeans_silhouette,", When no. Cluster:", i)
    
print("-"*70)

for i in range(2,7):
    agg = AgglomerativeClustering(n_clusters=i)
    agg_labels = agg.fit_predict(df_scaled)
    agg_silhouette = silhouette_score(df_scaled, agg_labels)
    print("Agglomerative clustering silhouette score:", agg_silhouette,", When no. Cluster:", i)
```

**Observation:**<br>
- K - Mean Clustering has better Silhouette Score than Agglomerative Clustering.
- As per the Slihouete Score also 2 and 4 can be the optimal number of clusters

### Question 4

```python
n_clusters_range = [2, 3, 4, 5, 6]
max_iter_range = [100, 300, 500]
init_range = ['k-means++', 'random']
algorithm_range = ['auto', 'full', 'elkan']

best_score_kmean = -1
for n_clusters in n_clusters_range:
    for max_iter in max_iter_range:
        for init in init_range:
            for algorithm in algorithm_range:
                kmeans = KMeans(n_clusters=n_clusters, max_iter=max_iter, init=init, algorithm=algorithm, random_state=42)
                kmeans_labels = kmeans.fit_predict(df_scaled)
                score = silhouette_score(df_scaled, kmeans_labels)
                param = {'n_clusters': n_clusters, 'max_iter': max_iter, 'init': init, 'algorithm': algorithm}
                print("Hyperparameter:",param,"Silhouette score: ",score)
                print("-"*70)
                if score > best_score_kmean:
                    best_score_kmean = score
                    best_params_kmean = {'n_clusters': n_clusters, 'max_iter': max_iter, 'init': init, 'algorithm': algorithm}
                    
print("Best hyperparameters:", best_params_kmean)
print("Best silhouette score:", best_score_kmean)
```

### Question 5

```python
n_clusters_range = [2, 3, 4, 5, 6]
metric_range = ['euclidean', 'manhattan', 'cosine']
linkage_range = ['ward', 'complete', 'average']

best_score_agg = -1
for n_clusters in n_clusters_range:
    for metric in metric_range:
        for linkage in linkage_range:
            try:
                agg = AgglomerativeClustering(n_clusters=n_clusters, metric=metric, linkage=linkage)
                agg_labels = agg.fit_predict(df_scaled)
                score = silhouette_score(df_scaled, agg_labels)
                param = {'n_clusters': n_clusters, 'metric': metric, 'method': linkage}
                print("Hyperparameter:",param,"Silhouette score: ",score)
                print("-"*70)
                if score > best_score_agg:
                    best_score_agg = score
                    best_params_agg = {'n_clusters': n_clusters, 'metric': metric, 'method': linkage}
            except:
                print(f"Error for n_clusters={n_clusters}, metric={metric}, method={linkage}")
                print("-"*70)

print("Best hyperparameters:", best_params_agg)
print("Best silhouette score:", best_score_agg)
```
