### Encoding

```python
def encoder(arr):
    encoding = {}
    for i in set(arr):
        if i not in encoding:
            encoding[i] = len(encoding)
    encoded = []
    for i in arr:
        encode = encoding[i]
        encoded.append(encode)
    return encoded
```

```python
train['Sex'] = encoder(train['Sex'])
train['Ticket'] = encoder(train['Ticket'])
train['Cabin'] = encoder(train['Cabin'])
train['Embarked'] = encoder(train['Embarked'])
```

### Model Evaluation

```python
def confusion_matrix(actual,predicted):
    TP = 0
    TN = 0
    FP = 0
    FN = 0
    for i in range(len(actual)):
        if (actual[i] == 1) and (predicted[i] == 1):
            TP += 1
    for i in range(len(actual)):
        if (actual[i] == 0) and (predicted[i] == 0):
            TN += 1
    for i in range(len(actual)):
        if (actual[i] == 0) and (predicted[i] == 1):
            FP += 1
    for i in range(len(actual)):
        if (actual[i] == 1) and (predicted[i] == 0):
            FN += 1
    return [[TN, FP],[FN,TP]]
```

```python
def accuracy(actual,predicted):
    TP = 0
    TN = 0
    FP = 0
    FN = 0
    for i in range(len(actual)):
        if (actual[i] == 1) and (predicted[i] == 1):
            TP += 1
    for i in range(len(actual)):
        if (actual[i] == 0) and (predicted[i] == 0):
            TN += 1
    for i in range(len(actual)):
        if (actual[i] == 0) and (predicted[i] == 1):
            FP += 1
    for i in range(len(actual)):
        if (actual[i] == 1) and (predicted[i] == 0):
            FN += 1
    if (TP+TN+FP+FN) == 0:
        accuracy = 0
    else:
        accuracy = (TP+TN)/(TP+TN+FP+FN)
    return accuracy
```

```python
def precision(actual,predicted):
    TP = 0
    TN = 0
    FP = 0
    FN = 0
    for i in range(len(actual)):
        if (actual[i] == 1) and (predicted[i] == 1):
            TP += 1
    for i in range(len(actual)):
        if (actual[i] == 0) and (predicted[i] == 0):
            TN += 1
    for i in range(len(actual)):
        if (actual[i] == 0) and (predicted[i] == 1):
            FP += 1
    for i in range(len(actual)):
        if (actual[i] == 1) and (predicted[i] == 0):
            FN += 1
    if (TP+FP) == 0:
        precision = 0
    else:
        precision = TP/(TP+FP)
    return precision
```

```python
def recall(actual,predicted):
    TP = 0
    TN = 0
    FP = 0
    FN = 0
    for i in range(len(actual)):
        if (actual[i] == 1) and (predicted[i] == 1):
            TP += 1
    for i in range(len(actual)):
        if (actual[i] == 0) and (predicted[i] == 0):
            TN += 1
    for i in range(len(actual)):
        if (actual[i] == 0) and (predicted[i] == 1):
            FP += 1
    for i in range(len(actual)):
        if (actual[i] == 1) and (predicted[i] == 0):
            FN += 1
    if (TP+FN) == 0:
        recall = 0
    else:
        recall = TP/(TP+FN)
    return recall
```

```python
def f1_score(actual,predicted):
    TP = 0
    TN = 0
    FP = 0
    FN = 0
    for i in range(len(actual)):
        if (actual[i] == 1) and (predicted[i] == 1):
            TP += 1
    for i in range(len(actual)):
        if (actual[i] == 0) and (predicted[i] == 0):
            TN += 1
    for i in range(len(actual)):
        if (actual[i] == 0) and (predicted[i] == 1):
            FP += 1
    for i in range(len(actual)):
        if (actual[i] == 1) and (predicted[i] == 0):
            FN += 1
    if (TP+FP) == 0:
        precision = 0
    else:
        precision = (TP/(TP+FP))
    if (TP+FN) == 0:
        recall = 0
    else:
        recall = TP/(TP+FN)
    if (precision+recall) == 0:
        f1_score = 0
    else:
        f1_score = (2*precision*recall)/(precision+recall)
    return f1_score
```

```python
print("\033[1m" +"\nSummary Of The Model"+ "\033[0m")
print(f"Confussion Matrix :{confusion_matrix(list(y_test),y_pred)}")
print("Accuracy : ",accuracy(list(y_test),y_pred))
print("Precision : ",precision(list(y_test),y_pred))
print("Recall : ",recall(list(y_test),y_pred))
print("F1 score : ",f1_score(list(y_test),y_pred))
```
