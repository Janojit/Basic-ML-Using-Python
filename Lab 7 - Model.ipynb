```python
import numpy as np
import pandas as pd
import math

import matplotlib.pyplot as plt
import seaborn as sns

from collections import OrderedDict
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
```

```python
def createClassificationModel(n_sam,n_feat,n_class): 
    X, y = make_classification(
        n_samples= n_sam,  
        n_features= n_feat , 
        n_informative=math.ceil(0.5*n_feat), 
        n_classes = n_class, 
        random_state = 42 
    )
    X_data = np.array(X)
    y_data = np.array(y)
    data = np.append(X_data ,y_data.reshape(-1,1),axis=1)
    df = pd.DataFrame(data)
    return df
```

```python
def renameCol(df):
    new_col = []
    i = 0
    for col in df.columns:
        if i ==  len(df.columns) - 1:
            new_col.append("target")
        else :
            new_col.append("f" + str(i))

        i += 1

    df.columns = new_col
```

```python
def stdScl(df,n_feat):
    sc = StandardScaler()
    for i in range(n_feat):
        df[df.iloc[:,:n_feat].columns[i]+"_StdScl"] = sc.fit_transform(df[[df.iloc[:,:n_feat].columns[i]]])
```

```python
def mmScl(df,n_feat):
    mm = MinMaxScaler()
    for i in range(n_feat):
        df[df.iloc[:,:n_feat].columns[i]+"_mmScl"] = mm.fit_transform(df[[df.iloc[:,:n_feat].columns[i]]])
```

```python
def logR(test_size, random_state,df,scl):
    X = df.drop(['target'], axis=1)
    y = df['target']
    if scl == "Standard Scaling":
        X_train, X_test, y_train, y_test = train_test_split(X.iloc[:,5:-5], y, test_size = test_size, random_state = random_state)
    if scl == "Min-Max Scaling":
        X_train, X_test, y_train, y_test = train_test_split(X.iloc[:,5:-10].join(X.iloc[:,-5:]), y, test_size = test_size, random_state = random_state)
    if scl == "None":
        X_train, X_test, y_train, y_test = train_test_split(X.iloc[:,:-10], y, test_size = test_size, random_state = random_state)
    model = LogisticRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    model = model
    acc = accuracy_score(y_test, y_pred)
    return model ,acc
```

```python
def SVM(test_size, random_state,df,scl):
    X = df.drop(['target'], axis=1)
    y = df['target']
    if scl == "Standard Scaling":
        X_train, X_test, y_train, y_test = train_test_split(X.iloc[:,5:-5], y, test_size = test_size, random_state = random_state)
    if scl == "Min-Max Scaling":
        X_train, X_test, y_train, y_test = train_test_split(X.iloc[:,5:-10].join(X.iloc[:,-5:]), y, test_size = test_size, random_state = random_state)
    if scl == "None":
        X_train, X_test, y_train, y_test = train_test_split(X.iloc[:,:-10], y, test_size = test_size, random_state = random_state)
    model = SVC()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    model = model
    acc = accuracy_score(y_test, y_pred)
    return model ,acc
```

```python
results_dataframe = pd.DataFrame(columns = ['Dataset','Scaling','Model', 'Test Size', 'Random State', 'Accuracy'])
random_states = [10,20,30,40,42,50,60]
test_sizes = [0.20,0.25,0.30]
scaling = ["Standard Scaling","Min-Max Scaling","None"]

for r_s in random_states:
    for ts in test_sizes:
        for scl in scaling:
            model, acc= logR(ts, r_s,df1,scl)
            result = {}
            result['Dataset'] = "Dataset 1"
            result['Scaling'] = scl
            result['Model'] = 'Logistic Regression'
            result['Test Size'] = ts
            result['Random State'] = r_s
            result['Accuracy'] = acc
            results_dataframe = results_dataframe.append(result, ignore_index = True)
```

```python
for r_s in random_states:
    for ts in test_sizes:
        for scl in scaling:
            model, acc= SVM(ts, r_s,df1,scl)
            result = {}
            result['Dataset'] = "Dataset 1"
            result['Scaling'] = scl
            result['Model'] = 'SVM'
            result['Test Size'] = ts
            result['Random State'] = r_s
            result['Accuracy'] = acc
            results_dataframe = results_dataframe.append(result, ignore_index = True)
```

```python
for r_s in random_states:
    for ts in test_sizes:
        for scl in scaling:
            model, acc= logR(ts, r_s,df2,scl)
            result = {}
            result['Dataset'] = "Dataset 2"
            result['Scaling'] = scl
            result['Model'] = 'Logistic Regression'
            result['Test Size'] = ts
            result['Random State'] = r_s
            result['Accuracy'] = acc
            results_dataframe = results_dataframe.append(result, ignore_index = True)
```
